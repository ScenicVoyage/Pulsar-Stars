```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Use x = Mean of the integrated profile            
# To predict y = Mean of the DM-SNR curve  

y = stars[" Mean of the DM-SNR curve"].values
X = stars.drop([" Standard deviation of the integrated profile", 
                " Excess kurtosis of the integrated profile", 
                " Skewness of the integrated profile", 
                " Standard deviation of the DM-SNR curve", 
                " Excess kurtosis of the DM-SNR curve", 
                " Skewness of the DM-SNR curve", 
                "target_class"], axis = 1) # everything else except for the target class column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train,y_train)
linear_pred = linear.predict(X_test)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(y_test, linear_pred)
ax.set_xlabel('Y Test')
ax.set_ylabel('Predicted Y')
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, linear_pred))
print('MSE:', metrics.mean_squared_error(y_test, linear_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, linear_pred)))
```

```python
sns.distplot((y_test-linear_pred), bins = 60)
```

```python
pd.DataFrame(linear.coef_, X.columns, columns = ['Coefficient']) 
```

```python
y = stars["target_class"].values
X = stars.drop(["target_class"], axis = 1) # everything else except for the target class column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train,y_train)
linear_pred = linear.predict(X_test)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(y_test, linear_pred)
ax.set_xlabel('Y Test')
ax.set_ylabel('Predicted Y')
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, linear_pred))
print('MSE:', metrics.mean_squared_error(y_test, linear_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, linear_pred)))
```

```python
sns.distplot((y_test-linear_pred), bins = 60)
```

```python
pd.DataFrame(linear.coef_, X.columns, columns = ['Coefficient'])
```

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

regression = LogisticRegression()
regression.fit(X_train, y_train)
regression_pred = regression.predict(X_test)
```

```python
print('Classification report for Logistic Regression: \n')

from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, regression_pred))
print(classification_report(y_test, regression_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

knn = KNeighborsClassifier(n_neighbors = 1) 

knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, 40), error_rate, color = 'blue', linestyle = 'dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors = 12)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print('WITH K = 12\n')
print(confusion_matrix(y_test, knn_pred))
print(classification_report(y_test, knn_pred))
```

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

svm = SVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
```

```python
print('Classification report for SVM Classification: \n')

print(confusion_matrix(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
```

```python
print('Classification report for Naive Bayes Classification: \n')
# assumes features are independent
print(confusion_matrix(y_test, nb_pred))
print(classification_report(y_test, nb_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
```

```python
print('Classification report for Decision Tree Classification: \n')

print(confusion_matrix(y_test, dt_pred))
print(classification_report(y_test, dt_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

rf = RandomForestClassifier(n_estimators = 100, random_state = 101)
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

```python
print('Classification report for Random Forest Classification: \n')

print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```
